```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\bowjo\Downloads\generalised_cognitive_dataset.csv")
df.head()
```

```python
df.info()        # Check datatypes and non-null values
df.describe()    # See statistics for all numeric columns
```

```python
sns.histplot(df['cognitive_score'], kde=True, color='blue')
plt.title("Distribution of Cognitive Score")
plt.xlabel("Cognitive Score")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(x='screen_time', y='cognitive_score', data=df)
plt.title("Screen Time vs Cognitive Score")
plt.xlabel("Screen Time (hrs/day)")
plt.ylabel("Cognitive Score")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
